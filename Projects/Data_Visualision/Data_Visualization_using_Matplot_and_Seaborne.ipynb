```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
# %matplotlib is used to specify the language to plot the chart right here in the code.
%matplotlib inline
```

```python
#Reading the CSV file
pokemon=pd.read_csv('pokemon.csv')
#Shap tells the total rows and columns
print(pokemon.shape)  
# Inspecting no. of rows from end
pokemon.tail(10)
```

# **Data Visualization Tools**
### **1) BAR CHARTS**
A bar chart is used to depict the distribution of a categorical variable. In a bar chart, each level of the categorical variable is depicted with a bar, whose height indicates the frequency of data points that take on that level. A basic bar chart of frequencies can be created through the use of seaborn's countplot function:

```python
plt.axes([0.05,1.1,0.9,0.9])
sb.countplot(data=pokemon,x = 'generation_id')
plt.axes([0.05,0.05,0.9,0.9])
sb.countplot(data = pokemon, y = 'type_1')
```

## **Specific colour to all bars**

```python
#Sea Borne provides a variety of Colors to Choose from the Color Palette
color_palete = sb.color_palette()
#Shows the list of all tuples(colors in RGB Format)
print(color_palete)
```

```python
#Assigns color at index 6 to Bar_color
Bar_color = sb.color_palette()[6]
#Assigning color to the Chart
sb.countplot(data = pokemon,x='generation_id',color=Bar_color)
```

## **Sorting the Chart**

```python
#Counting each generations of Pokemon
gen_count = pokemon['generation_id'].value_counts()
print(gen_count)
#Arranging in ascending order
gen_order = pokemon['generation_id'].value_counts().sort_values()
print(gen_order)
```

```python
# Sorting in Descending Order
bar_color = sb.color_palette()[0]
gen_order = pokemon['generation_id'].value_counts().index
sb.countplot(data = pokemon, x = 'generation_id', color = bar_color, order = gen_order)
```

```python
# Sorting in Ascending Order
bar_color = sb.color_palette()[1]
gen_order = pokemon['generation_id'].value_counts().sort_values().index
sb.countplot(data = pokemon, x = 'generation_id',color=bar_color,order = gen_order)
```

## **Additional Variations**
If you have a lot of category levels, or the category names are long, then you might end up with overcrowding of the tick labels as shown in below chart.


```python
bar_color = sb.color_palette()[4]
sb.countplot(data = pokemon, x = 'type_2', color = bar_color)
```

As you can see it is very difficult to read anything in the above chart to avoid this we can use the below methods.   
### **Method 1** 
By changing the tick axis

```python
bar_color = sb.color_palette()[4]
sb.countplot(data = pokemon, x = 'type_2', color = bar_color)
plt.xticks(rotation=90);
```

### **Method 2**
By Changing the axis itself

```python
bar_color = sb.color_palette()[4]
sb.countplot(data = pokemon, y = 'type_2', color = bar_color)
```

## **Counting Missing Data**
One interesting way we can apply bar charts is through the visualization of missing data. We can use pandas functions to create a table with the number of missing values in each column.

.isna() replaces Values with data to True and replaces Empty values with False as you can see below.  
Value of True is interpreted as 1 and Value of False is interpreted as 0.

```python
pokemon.isna().head()
```

Counting the total number of missing values using sum() method.   
For the pokemon Data set there are total of 402 empty values in Column type_1. 

```python
pokemon.isna().sum()
```

**Barplot** is used in Seaborne to plot custom Bar Charts.

```python
na_counts = pokemon.isna().sum()
base_color = sb.color_palette()[0]
sb.barplot(na_counts.index.values, na_counts, color = base_color)
plt.xticks(rotation=90);
```

#**2) Pie Chart**
A pie chart is a common univariate plot type that is used to depict relative frequencies for levels of a categorical variable. Frequencies in a pie chart are depicted as wedges drawn on a circle: the larger the angle or area, the more common the categorical value taken.

```python
# code for the pie chart seen above
counts = pokemon['generation_id'].value_counts()
# Using start angle to start with a vertical line
plt.pie(counts, labels = counts.index, startangle = 90,
        counterclock = True);
# Making sure the Radius are same from all sides
plt.axis('square');
```

# **3) Donut Chart**
A Donut Chart is similar to Pie Chart. The **only difference** is the Donut Chart is hollow from inside or say it has a hole in between just like a real donut.

A Donut Chart can be made using **wedgeprops** attribute inside pieplot.

```python
sorted_counts = pokemon['generation_id'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = True, wedgeprops = {'width' : 0.5});
plt.axis('square');
```

# **4) Histograms :**
Histograms are used to simply show data in continuous range, For ex- Chart for showing a person speed for 6 hrs / a day etc.       
values are grouped into continuous bins, and one bar for each bin is plotted depicting the number.

```python
bar_color = sb.color_palette()[8]
plt.hist(data = pokemon, x = 'defense',color = bar_color)
```

```python
# We can also decide the number of bins in the Chart.
plt.hist(data = pokemon, x = 'defense', bins = 15)
```

```python
H_bins = np.arange(0, pokemon['speed'].max()+1, 5)
plt.hist(data = pokemon, x = 'speed', bins = H_bins)
```

### **We don't want a graph that is not understandable ,For ex - the graph below on the right. We want a graph which is understandable like the graph on the left.**

```python
plt.figure(figsize = [10, 5]) # larger figure size for subplots

# histogram on left, example of too-large bin size
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
bin_edges = np.arange(0, pokemon['speed'].max()+1, +5)
plt.hist(data = pokemon, x = 'speed', bins = bin_edges);

# histogram on right, example of too-small bin size
plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
bin_edges = np.arange(0, pokemon['speed'].max()++1, 1)
plt.hist(data = pokemon, x = 'speed', bins = bin_edges);
```

## ***Creating Histogram using SEABORNE***
Seaborne gives the facality to create Charts in a manner that is so easy to understand. It has less functionality compared to Matplot but is very efficient while writing big code. Seaborne does most of the work itself related to charts.

```python
# Curve here is showing how the Pokemon speed is distributed.
sb.distplot(pokemon['speed']);
# Total area after integrating will be one.
```

```python
# Normal Histogram without the default Curve.
x = sb.distplot(pokemon['defense'], kde=False);
print(x)
```

Increasing the colour contrast using **hist_kws** property. 
Contrast can range from 0-1.

```python

bin_edges = np.arange(0, pokemon['speed'].max()+1, 5)
sb.distplot(pokemon['speed'], bins = bin_edges, kde = False,
            hist_kws = {'alpha' : 0.9});
```
